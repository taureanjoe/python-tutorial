```python
class Employee:
    
    def __init__(self, first, last, pay): # constructor / an init method
        # body of the constructor
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '-' + last + "@abaac.com"
        
    def fulname(self):
        return f"{self.first} {self.last}"
    
```

```python
class Employee:
    raise_amount = 1.04
    def __init__(self, first, last, pay): # constructor / an init method
        # body of the constructor
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '-' + last + "@abaac.com"
        
    def fulname(self):
        return f"{self.first} {self.last}"
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amount)
    

    
emp1 = Employee('Joe', 'Jacob', 50000)
emp2 = Employee('Arpit', "Agarwal", 90900)
```

```python
print(emp1.pay)
emp1.apply_raise()
```

```python
# polymorphism

class Ds:
    def dataisgood(self, name=' '):
        print("welcome to Dataisgood, " + name + "!")
        
obj = Ds()
obj.dataisgood()
obj.dataisgood("Arpit") # eg of overloading
```

```python
class Dscus8july(Ds):
    
    def dataisgood(self):
        print("Welcome to the batch")

obj = Dscus8july()
obj.dataisgood() # method overriding
```

```python
class Calculator:
    def addition(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b
    
    
class Calculate3(Calculator):
    def addition(self, a, b, c): # method overiding
        return a + b + c

    def subtract(self, a, b, c):
        return a - b - c
```

```python
sum1 = Calculate3()
sum1.addition(1,2,3)
```

```python
# encapsulation

class Payment:
    def __init__(self, price):
        self.__finalprice = price * price * 0.05
        
    def getfinalprice(self):
        return self.__finalprice
    
    def setfinalprice(self, discount):
        self.__finalprice = self.__finalprice - discount * 0.0001
```

```python
book = Payment(10)
```

```python
book.getfinalprice()
```

```python
book.setfinalprice()
```

```python
book.getfinalprice()
```
